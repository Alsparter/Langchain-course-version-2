```python
RAG:
1] Document loader 
2] Text Splitter 
3] Embeddings 
```

```python
1] Rag :
LLM + external knowledge [vector database]

Instead of relying  only on training data we, 
1] store documents
2] convert them into  embeddings
3] Retrive relevent chunks
4] inject them into the prompt
5] Generate grounded answer


Rag Architecture:

User Question
      ↓
Embedding Model
      ↓
Vector Store (FAISS / Chroma)
      ↓
Retriever
      ↓
LLM (with retrieved context)
      ↓
Final Answer
```

```python
Components of RAG:

1] Documents
2] Text Splitter
3] Embedding Model
4] Vector Store
5] Retriever
6] LLM


Types of RAG:

1] Naive RAG
2] Conversational RAG
3] Multi query RAG
4] Hybrid RAG
5] Agentic RAG

```

```python
from langchain.document_loaders import TextLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
from langchain.llms import HuggingFacePipeline
from langchain.prompts import PromptTemplate
from langchain.chains import RetrievalQA
from transformers import pipeline 


loader = TextLoader("sample.txt")
doc = loader.load()

splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
docs = splitter.split_documents(doc)

embed = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

vectorstore = FAISS.from_documents(docs,embed)
retriever = vectorstore.as_retriever(search_type="similarity", search_kwargs={"k":3})

pipe = pipeline("text-generation", model="distilgpt2", max_new_tokens=100)
llm = HuggingFacePipeline(pipeline=pipe)

qa_chain = RetrievalQA.from_chain_type(llm=llm, retriever=retriever, chain_type="stuff")

response = qa_chain.run("What is this document about ?")
print(response)
```

```python
LLMs cannot handle very long text  at once 

for eg
pdf =  200pages
one page = 2000 words
total = 4,00,000 words 

So we do 
load -> split -> embed -> store -> retrieve
```

```python
Text Splitter :
breaks large documents  into smaller  chunks that fit  into LLM context  windows

There are 2 concept of text splitter 

1] Chunk size:
How many characters per chunk
eg chunk_size=500

2] Chunk overlap:
How much content overlaps between chunks
eg chunk_overlap=100


Types :
1] RecursiveCharacterTextSplitter (most important)
2] CharacterTextSplitter
3] TokenTextSplitter
4] MarkdownHeaderTextSplitter (advanced)
```

```python
1] RecuersiveCharacterTextSplitter

It tries splitting in this order
    - paragrapgh \n\n
    - Sentence  \n
    - Spaces
    - Characters
So it preserve as much as possible

Original text (1000 char) suppose
after applying this splitter :;
chunk 1 : (0-500)
chunk 2: (400-900)
chunk 3: (800-1000)
```

```python
#RecursiveCharacterTextSplitter
from langchain_community.document_loaders import DirectoryLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter

#step 1; Load document
loader = DirectoryLoader("./documets", glob="*.pdf")
doc = loader.load()

#step 2: Text splitter 
text_splitter = RecursiveCharacterTextSplitter( chunk_size=500,chunk_overlap=100)
split_doc = text_splitter.split_documents(doc)

print("Total Chunks created: ",len(split_doc))
print("\n First Chunk \n")
print(split_doc[0].page_content)
print("\n metadata \n")
print(split_doc[0].metadata)
```

```python
3] TokenTextSplitter

Token is an word , part of word, puncation, sometimes even spaces

works:
chunk1 -> token 0-100
chunk2 -> token 80-180
chunk3 -> token 160-260
```

```python
from langchain_community.document_loaders import TextLoader
from langchain.text_splitter import TokenTextSplitter

#load 
loader = TextLoader("sample.txt")
doc = loader.load()

#split 
splitter = TokenTextSplitter(chunk_size=200, chunk_overlap=50)
split_doc = splitter.split_documents(doc)

print("Total chunks: ", len(split_doc))
print("\n First Chunks\n")
print(split_doc[0].page_content)
print("\n metadata \n")
print(split_doc[0].metadata)
```

```python
4] MarkdownHeaderTextSplitter
- Powerful splitter when you have documents 
    1)documentaion
    2) Technical notes
    3) README files 
    4) Knowledge bases
    5) Blogs writtenin markdown

splitting works :
chunk 1 -> Title
chunk  2 -> section
chunk 3 -> subsection
```

```python
#MarkdownTextSplitter based on text file
from langchain_community.document_loaders import TextLoader
from langchain.text_splitter import MarkdownHeaderTextSplitter

loader = TextLoader('sample.txt')
doc = loader.load()

#split
header_to_split_on = [
    ("#", "Header 1"), ("##", "Header 2")
]

markdown_splitter = MarkdownHeaderTextSplitter(headers_to_split_on=header_to_split_on)
split_doc = markdown_splitter.split_text(doc[0].page_content)

print("Total sections:", len(split_doc))
print("\n sections:")
print(split_doc[0].page_content)
print("\nmetadata\n")
print(split_doc[0].metadata)
```

```python
#MarkdownTextSplitter based on dcument folder 

from langchain_community.document_loaders import PyPDFLoader, DirectoryLoader
from langchain.text_splitter import MarkdownHeaderTextSplitter

loader = DirectoryLoader(path="./documets", glob="*.pdf", loader_cls=PyPDFLoader)
doc = loader.load()

print("Total Pdf pages loaded: ", len(doc))

full_text = ''

for d in doc:
    full_text += d.page_content + "\n"

print(f"\n  Full text : {full_text} \n")

header_to_split_on = [
    ("#", "Header 1"), ("##", "Header 2"), ("###", "Header 3")
]

splitter = MarkdownHeaderTextSplitter(headers_to_split_on=header_to_split_on)
split_doc = splitter.split_text(full_text)

print("-----------------------------------------------------------")
print("Total Sections Created: ", len(split_doc))

print("\nFirst section content: \n")
print(split_doc[0].page_content)
print("\n metadata \n")
print(split_doc[0].metadata)
```
