```python
from langchain_community.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS


#load Document
loader = PyPDFLoader("artificial_intelligence_tutorial.pdf")
doc = loader.load()

#split 
splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=100)
split_doc = splitter.split_documents(doc)
print("Total chunks:", len(split_doc))

#embeddings
embedding = HuggingFaceEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')

#vector store
vector_stores = FAISS.from_documents(split_doc, embedding)
```

```python
query = "what is Fuzzy logic"

results = vector_stores.similarity_search(query, k=2)
print("Top Results: \n")

for doc in results:
    print(doc.page_content)
    print("-"*50)
```

```python
result = vector_stores.similarity_search_with_score(query="What is fuzzy logic",k=2)
for doc, score in result:
    print("Score:", score)
    print(doc.page_content)
    print("-"*50)
```

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
from langchain.llms import HuggingFacePipeline
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline

loader = PyPDFLoader("artificial_intelligence_tutorial.pdf")
doc = loader.load()
splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=150)
doc_splitter = splitter.split_documents(doc)
embedding = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
vector_store = FAISS.from_documents(doc_splitter,embedding)
retriever = vector_store.as_retriever(search_type='similarity', search_kwargs={'k':1})

model_n = "distilgpt2"
model = AutoModelForCausalLM.from_pretrained(model_n)
tokenizer  = AutoTokenizer.from_pretrained(model_n)
pipe = pipeline('text-generation', model=model, tokenizer=tokenizer, max_new_tokens=300)
llm = HuggingFacePipeline(pipeline=pipe)

def navie_rag(query):
    result = retriever.get_relevant_documents(query)
    context = "\n\n".join([r.page_content for r in result])
    prompt = f"""
    Answer  the question using only the context below
    context:
    {context}

    question:
    {query}
    Answer:
    """
    response= llm(prompt)
    return response
```

```python
response = navie_rag("What is robotics")
print(response)
```

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
from langchain.llms import HuggingFacePipeline
from langchain.chains import ConversationalRetrievalChain
from langchain.memory import ConversationBufferMemory
from transformers import pipeline,AutoTokenizer, AutoModelForCausalLM

loader = PyPDFLoader("artificial_intelligence_tutorial.pdf")
doc = loader.load()
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
doc_splitter = text_splitter.split_documents(doc)
embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
vector_store = FAISS.from_documents(doc_splitter,embeddings)
retriver = vector_store.as_retriever(search_type="similarity", search_kwargs={"k":1})

model_name = "distilgpt2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
pipe = pipeline("text-generation", model=model, tokenizer=tokenizer, max_new_tokens=50, temperature=0.3)
llm = HuggingFacePipeline(pipeline=pipe)

memory = ConversationBufferMemory(memory_key="chat_history", return_messages=True, output_key="answer")
chain = ConversationalRetrievalChain.from_llm(llm=llm, memory=memory, retriever=retriver, return_source_documents=True)
```

```python
while True:
    question = input("You: ")

    if question.lower() == "exit":
        break

    result = chain.invoke({"question": question})

    print("\nAI:", result["answer"])
    print("\n" + "-"*60)
```

```python
RAG Types

1] Navie RAG
2] Conversational RAG
3] Multi query RAG
4] Hybrid RAG
5] Agentic RAG

6] Advance RAG
7] Reranker RAG
8] Graph RAG

3] Multi Query RAG :

Normal RAG  flow is user question -> Embed -> Similarity -> TOP k -> LLm response

MultiQuery RAG uses
- LLm to generate multiple variationsof the question
- Runs retrieval for each variations
- Merges all retrieved docs
- Removes duplicates
- Sends richer context to LLM

Simply it generate the asked question from user converted it to multiple forms and create multiple question within it and provide an single answer
MultiQuery RAG flow [ user question -> MultiQueryRetirever -> FAISS similarity search -> Merged unique documents  -> ConversationalRetrievalChain -> LLM generation

4] Hybrid RAG:

combines both FAISS sematic menaing and BM25 Exact  keyword matching

Flow of hybrid rag:
User Question -> Hybrid Retriever -> FAISS
									->  Merged Docs  -> LLm -> Answer
User Question -> Hybrid retrieval -> BM25


```

```python
##MultiQuery RAG Type 

from langchain_community.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain.retrievers.multi_query import MultiQueryRetriever
from langchain.chains import ConversationalRetrievalChain
from langchain_community.llms import HuggingFacePipeline
from transformers import pipeline,  AutoTokenizer, AutoModelForSeq2SeqLM

loader = PyPDFLoader("medicine.pdf")
doc = loader.load()
splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
split_docs = splitter.split_documents(doc)
print("Total Chunk Size :" len(split_docs))
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
vector_stores = FAISS.from_documents(split_docs, embeddings)


model_name = "google/flan-t5-base"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
pipe = pipeline("text-generation", model=model, tokenizer=tokenizer, max_new_tokens=200)
llm = HuggingFacePipeline(pipeline=pipe)

#multi query generation
base_retriever = vector_stores.as_retriever(search_type="mmr", search_kwargs={"k":2})
multi_query = MultiQueryRetriever.from_llm(retriever=base_retriever, llm=llm)

qa_chain = ConversationalRetrievalChain.from_llm(llm=llm,retriever= multi_query,  return_source_documents=False)

def ask_question(question: str):
    result = qa_chain.invoke({"question": question, "chat_history": []})
    return result["answer"]
    
```

```python
from  langchain_community.document_loaders import PyPDFLoader
from  langchain.text_splitter import RecursiveCharacterTextSplitter 
from  langchain.embeddings import HuggingFaceEmbeddings #C
from langchain.vectorstores import FAISS #c
from langchain.tools import Tool 
from langchain.agents import initialize_agent, AgentType
from langchain_community.llms import HuggingFacePipeline 
from transformers import pipeline, AutoTokenizer, AutoModelForCausalLM
import torch

loader = PyPDFLoader("")
doc = loader.load()
splitter = RecursiveCharacterTextSplitter(chunk_size=800, chunk_overlap=150)
doc_splittter = splitter.split_documents(doc)
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
vector_stores = FAISS.from_documents(doc_splittter, embeddings)
retriever = vector_stores.as_retriever(search_kwargs={"k":2})

model_name = ""
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, torch_dtype=torch.float32)
pipe = pipeline("text-generation", model=model, tokenizer=tokenizer, max_new_tokens=200, temperature=0.3)
llm = HuggingFacePipeline(pipeline=pipe)


def search_doc(query: str):
    docs = retriever.get_relevant_documents(query)
    return "\n\n".join([do.page_content for doc in docs])

retriever_tool = Tool(
    name="DocumentSearch", func=search_doc, description="Search the knowledge base for relevant  documents."
)

agent = initialize_agent(tools=[retriever_tool], llm=llm, agent=AgentType.ZERO_SHOT_REACT_DSCRIPTION, verbose=True)

def ask_agent(question: str):
    result = agent.run(question)
    return result 
```
