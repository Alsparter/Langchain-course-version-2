```python

# parameters to undertand
[ 1. Generatinal parameter ]
[ 2. Model configuration parameter ]
[ 3. tokenizer paramters]
[ 4. pipeline parameter ]

[ 1. Generatinal parameter ]

1] temperature :
Higher = more creative
lower = more accurate  and stable

2] max_tokens :
     control maximum output

3] top_p / top_k :
     Sampling control to adjust randomness
     - top p model chooses from top 90% likely words
     - top k  choose only from 50 most likelynext words


4] repetition_penalty :
     Penalize repating  same words/sentence

5] frequency_penalty :
     penalize words already used frequelntly

6] presence penalty :
	encourage  using new ideas not used before


[ 2. Model configuration parameter ]


1] model_name :
     Sepcify which model we are going to use

2] device_map :
     where model will run
     cpu, cuda, auto

3] dtype/precision :
     control memory and speed  of model
     float32, float16, bfloat16, int8, int4

4] trust_remote_code :
     allows loading custom model arch from hugging face  models
     if true

5] pad_token :
     padding token used during batching


[ 3. tokenizer paramters ]

1] truncation :
     cuts input if too long

2] padding :
     adds padding  to make input  length equal druing batch

3] max_length :
     Maximum length of input fed into model


[ 4. pipeline parameter ]

1] return_full_text :
     if false  only new generated text is returned

2] do_sample :
     enables sampling must be true when using tempreature  top_k, top_p

3] num_return_sequences :
     Generate multiple responses

```

```python
#exercise : text into embedding
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer('all-MiniLM-L6-v2')

S1 ="Python is a great language for AI."
S2 ="JavaScript is used for web development."


embedding1 = model.encode(S1)
embedding2 = model.encode(S2)

Similarity = util.cos_sim(embedding1, embedding2)
print("similarity score ", Similarity.item())
```

```python
# exercise : simple chatbot
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
from langchain.llms import HuggingFacePipeline
```

```python
# 1. choose open source
model_name = "distilgpt2"

# 2. load tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name)
tokenizer.pad_token = tokenizer.eos_token

# 3. Load model
model = AutoModelForCausalLM.from_pretrained(
	model_name, trust_remote_code =True
)

# 4. create text generation pipeline
pipe = pipeline(
	"text-generation",
	model=model,
	tokenizer=tokenizer,
	max_new_tokens=100,
	do_sample=True
)

# 5. Bind pipeline to langchain llm
llm = HuggingFacePipeline(pipeline=pipe)

# 6. Simple chat bot loop

print("chatbot is ready type quit to exit. \n")

while True:
     user_input = input('You: ')

     if user_input.lower() == 'quit':
          print("chatbot: goodbye!")
          break

     response = llm(user_input)
     print("chatbot:",response)
```
